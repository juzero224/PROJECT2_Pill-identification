```python
def f_get_dir_list(src_dir, dir_list):
    '''
    src_dir 안의 모든 서브 디렉토리를 출력
    예)
    dir_list = []
    f_get_dir_list(src_dir, dir_list)
    
    '''
    import os
    
    with os.scandir(src_dir) as entries:
        for entry in entries:
            if entry.is_dir():  # 디렉토리 이면
                entry_dir = os.path.join(src_dir, entry.name)
                
                # 리스트에 추가
                dir_list.append(entry_dir)
                f_get_dir_list(entry_dir, dir_list)
                
```

```python
def f_create_dir(dir_name):
    '''
    설명: 만들려는 디렉토리가 없으면 생성
    입력: 경로 + 새 디렉토리 이름
    출력: x
    예시: createDirectory('./test')
    '''
    import os
    
    try:
        if not os.path.exists(dir_name):
            os.makedirs(dir_name)
    except OSError:
        print("Error: Failed to create the directory.")
```

```python
import shutil
import os

src_dir = './alyac_img_label/google_drive'
png_dir = os.path.join(src_dir, 'png')
xml_dir = os.path.join(src_dir, 'xml')


# src_dir 안에 디렉토리 리스트 만들기
dir_list = []
f_get_dir_list(src_dir, dir_list)


# img, xml 디렉토리 생성
f_create_dir(png_dir)
f_create_dir(xml_dir)


# 확장자 별로 파일 복사
for entry_dir in dir_list:
    print(entry_dir)
    
    with os.scandir(entry_dir) as items:
        for item in items:
            # 파일이면
            if item.is_file():
                des_dir = ''
                ext = os.path.splitext(item)[1]  # 확장자 추출
                
                if ext == '.png':
                    des_dir = png_dir
                elif ext == '.xml':
                    des_dir = xml_dir
                    
                # 파일 복사
                if des_dir != '':
                    shutil.copy2(item, des_dir)

print('완료')
```
